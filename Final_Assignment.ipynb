```python
import yfinance as yf
import pandas as pd
import matplotlib_inline
```

```python
# Extract Tesla stock data from 2015 to 2023
tesla_data = yf.download('TSLA', start='2015-01-01', end='2023-01-01')

# Reset the index to move the Date index to a regular column
tesla_data_reset = tesla_data.reset_index()

# Display the first 5 rows
print(tesla_data_reset.head())
```

```python
# Fetch Tesla's financial data
tesla = yf.Ticker("TSLA")

# Extract the income statement (includes revenue)
income_statement = tesla.financials

# Extract the row for 'Total Revenue'
revenue = income_statement.loc['Total Revenue']

# Convert the revenue data into a DataFrame for easier manipulation
tesla_revenue = revenue.reset_index()

# Rename the columns for better readability
tesla_revenue.columns = ['Year', 'Revenue']

# Display the last 5 rows using the tail() function
print(tesla_revenue.tail())
```

```python
# Download GameStop stock data from 2015 to 2023
gme_data = yf.download('GME', start='2015-01-01', end='2023-01-01')

# Reset the index to move the Date index to a regular column
gme_data_reset = gme_data.reset_index()

# Display the first 5 rows of the gme_data dataframe
print(gme_data_reset.head())
```

```python
# Fetch GameStop's stock data using yfinance
gme = yf.Ticker("GME")

# Get the income statement, which includes the 'Total Revenue'
income_statement = gme.financials

# Extract 'Total Revenue' row
revenue = income_statement.loc['Total Revenue']

# Convert the revenue data into a DataFrame for easier manipulation
gme_revenue = revenue.reset_index()


# Display the last 5 rows using the tail() function
print(gme_revenue.tail())
```

```python
import matplotlib.pyplot as plt
# Define the make_graph function
def make_graph(stock_data, title):
    # Plot the stock's adjusted close price
    plt.plot(stock_data)
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.show()

# Call the make_graph function with Tesla's stock data and a title
make_graph(tesla_data, 'Tesla Stock Price (2015 - 2023)')
```
